Import Libraries

```python
import numpy as np
from scipy.stats import ttest_ind
import matplotlib.pyplot as plt
```

Input Sample Data


```python
sample1 = input("Enter data for sample 1, separated by spaces: ")
sample1 = [float(x) for x in sample1.split()]

sample2 = input("Enter data for sample 2, separated by spaces: ")
sample2 = [float(x) for x in sample2.split()]
```

Calculate CV

```python
cv1 = np.std(sample1) / np.mean(sample1) * 100
cv2 = np.std(sample2) / np.mean(sample2) * 100

# Print the CV for each dataset
print("CV for Manual:", cv1)
print("CV for Automated:", cv2)

# Compare the CVs and print the result
if abs(cv1 - cv2) < 10:
    print("The two datasets do not deviate a lot.")
else:
    print("The two datasets deviate significantly.")
```

```python
# Calculate means
mean1 = np.mean(sample1)
mean2 = np.mean(sample2)

# Create plot
fig, ax = plt.subplots()
ax.scatter(range(len(sample1)), sample1, label="Manual")
ax.scatter(range(len(sample2)), sample2, label="Automated")
ax.axhline(mean1, color='r', linestyle='--', label="Mean of Manual")
ax.axhline(mean2, color='b', linestyle='--', label="Mean of Automated")

# Add legend and labels
ax.legend()
ax.set_xlabel("Sample")
ax.set_ylabel("Value")
ax.set_title("Deviation between Manual and Automated")

# Show plot
plt.show()
```

T-test

```python
t_stat, p_value = ttest_ind(sample1, sample2)

print("t-statistic:", t_stat)
print("p-value:", p_value)

plt.hist(sample1, alpha=0.5, label='Manual')
plt.hist(sample2, alpha=0.5, label='Automated')
plt.axvline(np.mean(sample1), color='blue', linestyle='dashed', linewidth=2)
plt.axvline(np.mean(sample2), color='orange', linestyle='dashed', linewidth=2)
plt.legend()
plt.title("t-test results: t = {:.2f}, p = {:.4f}".format(t_stat, p_value))
plt.show()
```
